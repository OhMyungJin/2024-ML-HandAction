```python
import moviepy.editor as mp
import os

# 변환할 동영상들이 있는 디렉토리 경로
input_directory = "/content/drive/MyDrive/ML(hello)"
output_directory = "/content/drive/MyDrive/ML(hello)_Output"
target_duration = 2  # 목표 길이 (초)
target_extension = "mov"  # 목표 확장자

if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# 디렉토리 내의 모든 동영상 파일을 가져오기
for filename in os.listdir(input_directory):
    if filename.endswith((".mp4", ".avi", ".mov", ".mkv")):  # 지원되는 확장자 리스트
        filepath = os.path.join(input_directory, filename)
        video = mp.VideoFileClip(filepath)

        # 목표 길이에 맞게 동영상 자르기
        if video.duration > target_duration:
            video = video.subclip(0, target_duration)
        elif video.duration < target_duration:
            # 동영상 길이가 목표 길이보다 짧을 경우, 반복해서 맞추기
            loops = int(target_duration // video.duration) + 1
            video = mp.concatenate_videoclips([video] * loops).subclip(0, target_duration)

        # 새로운 파일 이름과 경로 설정
        new_filename = os.path.splitext(filename)[0] + "." + target_extension
        output_path = os.path.join(output_directory, new_filename)

        # 동영상을 목표 확장자로 저장
        video.write_videofile(output_path, codec='libx264')

print("동영상 변환이 완료되었습니다.")
```
